### Subject Level Data

- Data from Vessel (2018): 5 Types of Image; Rating + Keypress Assessment

```python
vessel_subject_data = load_response_data('vessel', average = False)
```

 - Original Oasis Data (Kurdi, 2018): 900 Images; 4 Types of Image; Arousal + Valence Ratings
 - Data from Brielmann (2019): Beauty Ratings for all 900 Oasis Images

```python
oasis_subject_data = load_response_data('oasis', average = False)
```

```python
count_data = (oasis_subject_data.melt(id_vars = ['subject','image_name','image_type'], 
                                      value_vars = ['arousal','beauty','valence'],
                                      var_name = 'measurement', value_name = 'rating')
              .query('~rating.isnull()', engine = 'python')
              .groupby(['image_name', 'measurement'])['subject'].nunique()).reset_index()
```

```python
count_data.groupby('measurement')['subject'].agg([np.min, np.mean, np.max])
```
