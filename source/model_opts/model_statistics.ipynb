### VISSL Models

```python
### note the use of the extract_links argument in read_html from pandas requires pandas>=1.5.1
def extract_hyperlink(cell):
    if cell[1] is not None and any(x in cell[1] for x in ['download','dl']):
        return cell[1]
    else:
        return cell[0]

vissl_tables = pd.concat(pd.read_html('https://github.com/facebookresearch/vissl/blob/main/MODEL_ZOO.md',
                            extract_links='body')).applymap(extract_hyperlink, na_action = 'ignore')

target_columns = ['Method', 'Model', 'PreTrain dataset', 'URL', 'ImageNet top-1 acc.',
                  'ImageNet top-1 fine-tuned acc.', 'ImageNet top-1 linear acc.', 'ImageNet k-NN acc.']

vissl_tables = vissl_tables[target_columns]
vissl_tables.columns = ['method', 'model', 'train_data','weights_url',
                        'imagenet_top1_supervised','imagenet_top1_linear', 
                        'imagenet_top1_finetuned', 'imagenet_top1_knn']
```
